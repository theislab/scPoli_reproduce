```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.colors import LogNorm
```

```python
plt.rcParams['figure.dpi'] = 500
%config InlineBackend.figure_formats = ['pdf']
```

```python
h = np.array([10, 25, 50, 100, 250, 500])
n = np.array([10, 50, 100, 500, 1000, 5000])
```

```python
def emb(H, N):
    return (N*H) / (N + H)
```

```python
h = np.array([10, 50, 100, 500, 1000])
n = np.array([10, 50, 100, 500, 1000, 5000])
H, N = np.meshgrid(h, n)
Z = emb(H, N)

fig, ax = plt.subplots(1, 1)
# Plot the surface.
surf = ax.imshow(Z)
```

```python
fig, ax = plt.subplots(1, 1)
sns.heatmap(
    Z.astype(int), 
    norm=LogNorm(), 
    ax=ax, 
    annot=True, 
    fmt='d', 
    cbar_kws={'label': 'Upper bound of E dimensionality'}
)
ax.set_xticklabels(h.astype(int))
ax.set_yticklabels(n.astype(int))
ax.set_xlabel('Hidden layer size (H)')
ax.set_ylabel('Number of samples (N)')
ax.invert_yaxis()
```
