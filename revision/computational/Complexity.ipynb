```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.dpi'] = 300
%config InlineBackend.figure_formats = ['pdf']
plt.rcParams['axes.facecolor'] = (1, 1, 1, 0)
```

```python
results = pd.DataFrame({
    'memory (kB)': [22116032, 23308808, 23748592, 28008932, 70845128],
    'n_cells': [0, 1000, 10000, 100000, 1000000],
    'time': [5, 16, 72, 637, 6831]
})
results['net memory'] = results['memory (kB)'] - results['memory (kB)'].loc[0]
```

```python
results_sparse = pd.DataFrame({
    'memory (kB)': [3405788, 4539876, 4562272, 4759252, 6811792],
    'n_cells': [0, 1000, 10000, 100000, 1000000],
    'time': [8, 26, 179, 1631, 17816]
})
results_sparse['net memory'] = results_sparse['memory (kB)'] - results_sparse['memory (kB)'].loc[0]
```

```python
results = results.assign(data='dense')
results_sparse = results_sparse.assign(data='sparse')
results = pd.concat([results, results_sparse])
results.reset_index(inplace=True)
```

```python
results['net memory (GB)'] = results['net memory'] * 1e-6
results['memory (GB)'] = results['memory (kB)'] * 1e-6
```

```python
results = results[results['n_cells'] != 0]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 3))
g = sns.barplot(x='n_cells', y='time', hue='data', data=results, ax=ax[0])
ax[0].set_xticklabels(['1K', '10K', '100K', '1M'])
ax[0].set_xlabel('Dataset size')
ax[0].set_ylabel('Time')
#ax[0].get_legend().remove()
g.set_yscale('log')
g = sns.barplot(x='n_cells', y='net memory (GB)', hue='data', data=results, ax=ax[1])
g.set_yscale('log')
ax[1].set_xticklabels(['1K', '10K', '100K', '1M'])
ax[1].set_xlabel('Dataset size')
ax[1].set_ylabel('Memory')
ax[1].get_legend().remove()
fig.tight_layout()
sns.despine(offset=5)
```
