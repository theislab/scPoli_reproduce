```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scanpy as sc
import seaborn as sns
```

```python
adata_latent_source = sc.read('haniffa_latent.h5ad')
```

```python
sc.pp.pca(adata_latent_source)
sc.pp.neighbors(adata_latent_source)
sc.tl.umap(adata_latent_source)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sc.pl.umap(adata_latent_source, color='Site', show=False, ax=ax)
sns.despine()
fig.savefig('../figures/haniffa_int_umap_site.png', dpi=600, bbox_inches='tight')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 5))
sc.pl.umap(adata_latent_source, color='full_clustering', show=False, ax=ax)
sns.despine()
fig.savefig('../figures/haniffa_int_umap_ct.png', dpi=600, bbox_inches='tight')
```

```python
adata_latent_source.write('haniffa_latent.h5ad')
```
